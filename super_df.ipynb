```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from os.path import join as oj

root_folder = '/Users/yang/Documents/Wilbrecht_Lab/sleap_video/preprocessed_data'

def plot_two_column(df, animal, session, x, y):
    plt.figure(figsize=(10, 4))
    sns.lineplot(data=df, x=x, y=y)
    plt.title(f'{animal} {session}')
    plt.xlabel(f'{x}')
    plt.ylabel(f'{y}')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def plot_trajectory(df, animal, session):
    plt.figure(figsize=(10, 6))
    palette = sns.color_palette('deep', n_colors=len(df['trial'].unique()))
    for color, (trial_num, group) in zip(palette, df.groupby('trial')):
        sns.lineplot(
            data=group,
            x='warped Head x',
            y='warped Head y',
            color=color,
            linewidth=0.5,
            legend=False,
            sort=False
        )
       
    plt.title(f'{animal} {session}')
    plt.show()
```
